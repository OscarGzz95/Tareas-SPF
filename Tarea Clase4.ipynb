1. Imagine una rutina contable utilizada en una librera. Funciona en una lista con sublistas, que se ven as:

Escriba un programa en Python, que devuelve una lista con tuplas de tamao 2. Cada tupla consiste en el nmero de pedido y el producto del precio por artculo y la cantidad. El producto debe aumentarse en 10, -  si el valor de la orden es inferior a 100,00 . 
Escribe un programa Python usando **lambda y map**.



```python
import numpy as np
import time,math 
import matplotlib.pyplot as plt
```

```python
orden = [ ["34587", "Learning Python, Mark Lutz", 4, 40.95],
          ["98762","Programming Pythin, Mark Lutz", 5, 56.80],
          ["77226","Head First Python, Paul Barry", 3, 32.95],
          ["88112", "Einfhrung in Python, Bernd Klein", 3, 24.99]]

tup = lambda x, y: x*y

def increase(value):
    if(value < 100):
        value+=10
    return value

x = list(map(increase,[tup(orden[i][2], orden[i][3]) for i in np.arange(len(orden))]))

result = list(map(lambda x,y:(x[0],y), orden, x))
print(result)
```

2. La misma librera, pero esta vez trabajamos en una lista diferente. Las sublistas de nuestras listas se ven as: 
[nmero de orden, (nmero de artculo, cantidad, precio por unidad), ... (nmero de artculo, cantidad, precio por unidad)] 
> orders = [ [1, ("5464", 4, 9.99), ("8274",18,12.99), ("9744", 9, 44.95)], 
>	       [2, ("5464", 9, 9.99), ("9744", 9, 44.95)],
>	       [3, ("5464", 9, 9.99), ("88112", 11, 24.99)],
>          [4, ("8732", 7, 11.99), ("7733",11,18.99), ("88112", 5, 39.95)] ]

Escriba un programa que devuelva una lista de dos tuplas (nmero de pedido, cantidad total de pedido).

```python
from functools import reduce
```

```python
orders = [ [1, ("5464", 4, 9.99), ("8274",18,12.99), ("9744", 9, 44.95)],
          [2, ("5464", 9, 9.99), ("9744", 9, 44.95)],
          [3, ("5464", 9, 9.99), ("88112", 11, 24.99)],
          [4, ("8732", 7, 11.99), ("7733",11,18.99), ("88112", 5, 39.95)] ]

min_orden = 100
lista1 = list(map(lambda x: [x[0]] + list(map(lambda y: y[1]*y[2] if y[1]*y[2]>= min_orden else y[1]*y[2] +10, x[1:])), orders))
result = list(map(lambda x: [x[0]] + [reduce(lambda a,b: a+b,x[1:])], lista1))
print(result)
```
